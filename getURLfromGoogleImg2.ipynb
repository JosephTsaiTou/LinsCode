## 定義 Google 圖片網址下載爬蟲
### 用 chrome 瀏覽器開啟圖片搜尋頁面，拉到底後按下"顯示更多結果"按鈕，再下拉到底，將所有圖片網址存放置 "E:/Group_1/ImgURLs/classname.txt" 中

# executable_path 請改成自己電腦的設定

```python
import sys
from time import sleep

from splinter import Browser
from bs4 import BeautifulSoup

def getURL(classname):

    url = "https://www.google.com.tw/search?hl=zh-TW&site=imghp&tbm=isch&source=hp&biw=1450&bih=806&q="

    #executable_path = {'executable_path':'<C:/Program Files (x86)/chromedriver/>'}
    #browser = Browser('chrome', **executable_path)
    browser = Browser('chrome', executable_path='E:/ETL/chromedriver.exe')

    browser.visit(url + classname)
    sleep(3)

    for i in range(1,15):
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        sleep(3)
        if i == 5:
            browser.find_by_id('smb').click()
            sleep(3)

    while True:
        is_loaded = browser.evaluate_script("document.readyState")
        #print(is_loaded)
        if is_loaded == 'complete':
            break
        else:
            sleep(1)

    soup = BeautifulSoup(browser.html, 'lxml')

    count = 0

    for tag in soup.select('div.rg_meta'):
        try:
            #print(str(count))
            str1 = tag.text
            str2 = str1.replace('false', '"haha"')
            str3 = str2.replace('true', '"hahaha"')
            dict3 = eval(str3)
            dict3['ou']
            #print(dict2['ou'])
            count += 1
            with open('E:/Group_1/ImgURLs/' + classname +'.txt', 'a', encoding='utf8') as f2:
                f2.write(dict3['ou'] + "\n")
        except:
            print(str(count) + " ERROR!!!\n" + tag.text + "\n")
            tp, val, tb = sys.exc_info()
            print(sys.stderr, '%s: %s' % (str(tp), str(val)))
        
#         錯誤輸出
#         tp exception 的 type
#         val 是 raise 時，第二個參數
#         tb 則保留了 exception 發生時的呼叫狀態和堆疊內容 

    print("-" * 70)
    print(classname +'.txt has ' + str(count) + ' URLs.')

    browser.quit()
```

## 建立欲搜尋字串的 List，用 for 取出後呼叫方法

```python
name_list = []

with open('E:/Group_1/ClassnameList/' + os.listdir("E:/Group_1/ClassnameList/")[0]) as f:
    while True:
        line = f.readline()
        if not line:
            break
        name_list.append(line.strip())
```

```python
for classname in name_list:
    getURL(classname)
```
